```python
Cough_dataframe = pd.read_csv("./cough.csv")
```

```python
Cough_dataframe.head(15)
```

```python
Cough_dataframe["216"].value_counts()
```

```python
X = Cough_dataframe.iloc[:,:216]
y = Cough_dataframe.iloc[:,216]
```

```python
C_range = np.logspace(-1, 10, 12)
gamma_range = np.logspace(-1, 20, 22)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2)
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv, verbose=3, n_jobs=-1)
grid.fit(X, y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
classification_report(y,grid.predict(X))
```
